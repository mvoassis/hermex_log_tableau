```python
import pandas as pd
```

```python
pedidos = pd.read_csv('base-de-dados-Hermex - Pedidos.csv')
pedidos.head()
```

```python
pedidos.UF = [uf[-2:] for uf in pedidos.UF]
pedidos.head()
```

```python
states = {
    'AC': 'Acre',
    'AL': 'Alagoas',
    'AP': 'Amapá',
    'AM': 'Amazonas',
    'BA': 'Bahia',
    'CE': 'Ceará',
    'DF': 'Distrito Federal',
    'ES': 'Espírito Santo',
    'GO': 'Goiás',
    'MA': 'Maranhão',
    'MT': 'Mato Grosso',
    'MS': 'Mato Grosso do Sul',
    'MG': 'Minas Gerais',
    'PA': 'Pará',
    'PB': 'Paraíba',
    'PR': 'Paraná',
    'PE': 'Pernambuco',
    'PI': 'Piauí',
    'RJ': 'Rio de Janeiro',
    'RN': 'Rio Grande do Norte',
    'RS': 'Rio Grande do Sul',
    'RO': 'Rondônia',
    'RR': 'Roraima',
    'SC': 'Santa Catarina',
    'SP': 'São Paulo',
    'SE': 'Sergipe',
    'TO': 'Tocantins'
}
```

```python
pedidos.ESTADO = pedidos.UF.map(states)
pedidos.head()
```

```python
pedidos.to_csv('/content/base-de-dados-Hermex - Pedidos prep.csv', header=True, index=False)
```
